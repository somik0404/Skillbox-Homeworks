На выходе программа должна выдавать, сколько дней пользователю потребуется для того, чтобы подготовиться пробежать целевое количество километров.

*Ограничение: нельзя пользоваться функцией `ceil()` из модуля `math` и её аналогами.*

````python
days = 1
distance = int(input())
first_day = int(input())
percent = int(input())
while first_day <= distance:
    first_day = first_day * (1 + percent / 100)
    days += 1
print(days)
````

### 17

Напишите программу, которая выводит число пар одинаковых элементов в списке. При этом, если два элемента образовали пару, они не могут вступить в пару с другими элементами. Программа должна запрашивать на вход слова, каждое с новой строки, пока пользователь не введёт пустую строку. 
<br><br>
*Формат ввода:* <br>
Алиса<br>
Алиса<br>
Алиса<br>
<br><br>

*Формат вывода:* <br>
1

**Рекомендации к выполнению задания:** попробуйте использовать проверку на включённость.

````python
list_of_names = []
names_counter = {}
couples = 0
while True:
    x = input()
    if x == '':
        break
    else:
        list_of_names.append(x)
for name in list_of_names:
    if name in names_counter:
        names_counter[name] += 1
    else:
        names_counter[name] = 1
for k in names_counter:
    couples += names_counter[k] // 2
print(couples)
````
